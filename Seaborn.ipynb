# Visualization with Seaborn

```python
#import required libraries
import seaborn as sns
sns.set()
sns.set(style = 'darkgrid')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# loading dataset
BigMart = pd.read_csv(r'C:\Users\ashab\Downloads\Pandas_Part_1\Pandas Part 1\bigmart_data.csv')
BigMart = BigMart.dropna(how = 'any')

# multiply item visibility by 100 to increase size
BigMart['Visibility_Scaled'] = BigMart['Item_Visibility'] * 100
BigMart.head()
```

# Creating Basic plots

# 1. Line Chart

```python
sns.lineplot(x = 'Item_Weight', y = 'Item_MRP', data= BigMart[:50] )
```

# 2. Bar chart

```python
sns.barplot(x = 'Item_Type', y = 'Item_MRP', data = BigMart[:5])
```

# 3. Histogram

```python
sns.distplot(BigMart['Item_MRP'])
```

# 4. Box chart

```python
sns.boxplot(BigMart["Item_Outlet_Sales"], orient = 'vertical')
```

# 5. Violin plot

```python
sns.violinplot(BigMart['Item_Outlet_Sales'], orient = 'vertical', color = 'aqua')
```

# Category Wise Subplot

```python
sns.relplot(x = "Item_Weight", y = "Item_Visibility", hue = "Outlet_Size", style = "Outlet_Size", col = "Outlet_Size", data = BigMart[:50]);
```

# Categorical plots

```python
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = 'strip', data = BigMart[:250])
```

```python
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "swarm", data = BigMart[:250])
```

```python
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "box", data = BigMart)
```

```python
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "violin", data = BigMart)
```

```python
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "bar", data = BigMart)
```
