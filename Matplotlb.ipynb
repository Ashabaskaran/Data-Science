# Visualization with Matplotlib

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Display plots in notebook itself
%matplotlib inline
```

# Make a simple plot

```python
height = [155, 165, 175, 185, 195]
weight = [70, 80, 90, 100, 110]

plt.plot(height, weight)
```

```python
plt.plot(height, weight)

#add title
plt.title("Difference between height and weight")

#label x axis
plt.xlabel("Height")

#label y axis
plt.ylabel("weight") 

```

```python
calories_burnt = [100, 90, 80, 70, 60]

#Figure size in inches
plt.figure(figsize = (15, 5))

plt.plot(calories_burnt, 'y--')
plt.plot(weight)

#add legend
plt.legend(labels = ['Calories_burnt', 'Weight'], loc = 'lower right')

#set labels for each of these person
plt.xticks(ticks = [0, 1, 2, 3], labels = ['p1', 'p2', 'p3', 'p4'])
```

```python
#create two plots
fig, ax = plt.subplots (nrows = 1, ncols = 2, figsize = (6, 6), sharey = True)

#plot on 0 row and 0 column
ax[0].plot(calories_burnt, 'go')

#plot on 0 row and coumn 1
ax[1].plot(weight)

#set titles for subplots
ax[0].set_title("Calories_Burnt")
ax[1].set_title("Weight")

#set ticks for each of this person
ax[0].set_xticks(ticks = [0, 1, 2, 3]);
ax[1].set_xticks(ticks = [0, 1, 2, 3]);

#set labrls for each of these person
ax[0].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4']);
ax[1].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4']);
```

```python
#read the data set
BigMart = pd.read_csv(r'C:\Users\ashab\Downloads\Pandas_Part_1\Pandas Part 1\bigmart_data.csv')
#drop the null values
BigMart = BigMart.dropna(how = 'any')
BigMart.head()
```

```python
price_by_item = BigMart.groupby('Item_Type').Item_MRP.mean()[:10]
plt.figure(figsize =[18, 8])
plt.plot(price_by_item)
x = price_by_item.index.tolist()
y = price_by_item.values.tolist()

plt.xlabel("Item type")
plt.ylabel("Mean Price")
plt.title("Mean price for each item")

#set xticks
plt.xticks(labels=x, ticks=np.arange(len(x)))
plt.plot(x, y)
```

# Bar Chart

```python
#Sales by outlet size
sales_by_outlet_size = BigMart.groupby('Outlet_Size').Item_Outlet_Sales.mean()

#Sort by sales
sales_by_outlet_size.sort_values(inplace = True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

#Set axis labels
plt.xlabel('Outlet Size')
plt.ylabel("Sales")

#Set Title
plt.title('Mean sales for each outlet size')

#Set xticks
plt.xticks(labels =x, ticks = np.arange(len(x)))

plt.bar(x, y, color = ['red', 'orange', 'magenta'])
```

# Histogram

```python
#add title
plt.title('Item MRP(price) distribution')

#Add labels
plt.xlabel('Item_MRP')
plt.ylabel('Frequency')

#Plot Histogram
plt.hist(BigMart['Item_MRP'], bins = 30, color = 'Orange')
```
